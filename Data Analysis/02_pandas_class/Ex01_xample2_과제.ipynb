# 대학별 대통령의 평균 취임나이

```python
df3 = df[['College','Age at inauguration']] # 데이터에서 컬럼뽑기

df_pre2 = df3.groupby('College').mean() #평균치 값 df_pre2에 넣기

df_pre2.plot(kind='barh') # plot
```
