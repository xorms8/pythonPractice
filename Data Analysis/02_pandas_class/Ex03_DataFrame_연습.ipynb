#### 행에서 데이타 추출

- 야채, 딸기, 봉투행을 추출하여 매출순으로 정렬하기 (sort_values()이용)

- 야채부터 딸기행까지 추출

```python
df.loc[['야채','딸기','봉투']].sort_values('매출', ascending=False)

#df['바나나':'딸기'] 에러발생

df.loc['야채':'딸기','가격'] # 야채부터 딸기행, 그리고 가격컬럼표시 ! 
```

#### 조건으로 검색

- 가격이 10000 초과 상품 출력
- 가격이 10000에서 20000 사이 상품 출력

```python
#가격이 10000초과 상품 출력
df[df['가격'] > 10000]
```

```python
#가격이 10000에서 20000사이 상품 출력
df[(df['가격'] >= 10000) & (df['가격'] <= 20000)]
```

```python
price=df['가격']
print(price)
for i in price:
    if i > 10000:
        print(i)
```
